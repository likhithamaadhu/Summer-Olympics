# 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
d=df.values
```

```python
l=[]

for sport in df['Sport'].unique():
    v=0
    for i in range(len(d)):
        if((d[i][2]==sport)and(d[i][8]=='Gold')):
                v+=1
    l.append([sport,v])
pd.DataFrame(l,columns = ['sport','no.of.goldmedals']).sort_values(by='no.of.goldmedals', ascending=False).head().plot(x = 'sport', y = 'no.of.goldmedals', kind = 'bar', figsize = (5,5))
```

## 3.Which sport is having most number of medals so far?(Top 5)

```python
l=[]

for sport in df['Sport'].unique():
    v=0
    for i in range(len(d)):
        if((d[i][2]==sport)and((d[i][8]=='Gold')|(d[i][8]=='Silver')|(d[i][8]=='Bronze'))):
                v+=1
    l.append([sport,v])

pd.DataFrame(l,columns = ['sport','no.of.medals']).sort_values(by='no.of.medals', ascending=False).head().plot(x = 'sport', y = 'no.of.medals', kind = 'bar', figsize = (5,5))
```
